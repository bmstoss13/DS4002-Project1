```python
"""
This file produces summary statistics and visualizations of each variable in our final dataset, as seen in the Data Appendix.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"/Users/Nata/Downloads/preprocessed_data.csv")
df.head()
```

```python
# Check for missing values 

print("\nMissing Values:")
print(df.isnull().sum())
print("\nLength of Columns:")
for col in df:
    print(col,":", len(df[col]))
```

```python
# Frequency table for 'Spam or Not Spam' column

spam_freq_table = df['Spam or Not Spam'].value_counts()
print(spam_freq_table)
```

```python
# Frequency distribution graph for 'Spam or Not Spam' column

df['Spam or Not Spam'].value_counts().plot(kind='bar', xlabel='Email Type', ylabel='Frequency', edgecolor = 'black')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Summary statistics for 'Text Length' column

df['Text Length'].describe()
```

```python
# Histogram of 'Text Length' column

plt.figure(figsize=(8, 5))
sns.histplot(df["Text Length"], bins=30, kde=True)
plt.title("Distribution of Email Lengths")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")
plt.show()
```

```python
# Summary statistics for 'Word Count' column

df['Word Count'].describe()
```

```python
# Histogram of 'Word Count' column

plt.figure(figsize=(8, 5))
sns.histplot(df["Word Count"], bins=30, kde=True)
plt.title("Distribution of Email Word Count")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.show()
```

```python
# Summary statistics for 'Sentence Count' column

df['Sentence Count'].describe()
```

```python
# Histogram of 'Sentence Count' column

plt.figure(figsize=(8, 5))
sns.histplot(df["Sentence Count"], bins=30, kde=True)
plt.title("Distribution of Email Sentence Count")
plt.xlabel("Sentence Count")
plt.ylabel("Frequency")
plt.show()
```
